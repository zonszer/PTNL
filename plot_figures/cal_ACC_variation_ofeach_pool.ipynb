```python
text_before="""
pool_id: 0, unc_max: 3.8594, pool ACC: 2/2, pool_capacity: 2/5
pool_id: 1, unc_max: 3.4570, pool ACC: 1/4, pool_capacity: 4/5
pool_id: 2, unc_max: 2.5059, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 3, unc_max: 1.9092, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 4, unc_max: 0.0000, pool ACC: 0/0, pool_capacity: 0/5
pool_id: 5, unc_max: 1.7373, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 6, unc_max: 3.7695, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 7, unc_max: 3.3262, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 8, unc_max: 3.1289, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 9, unc_max: 2.3516, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 10, unc_max: 3.7070, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 11, unc_max: 0.0000, pool ACC: 0/0, pool_capacity: 0/5
pool_id: 12, unc_max: 3.6855, pool ACC: 1/5, pool_capacity: 5/5
pool_id: 13, unc_max: 2.1523, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 14, unc_max: 2.8086, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 15, unc_max: 3.6953, pool ACC: 1/1, pool_capacity: 1/5
pool_id: 16, unc_max: 2.6270, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 17, unc_max: 3.7637, pool ACC: 1/5, pool_capacity: 5/5
pool_id: 18, unc_max: 2.6641, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 19, unc_max: 3.4707, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 20, unc_max: 0.6353, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 21, unc_max: 0.0000, pool ACC: 0/0, pool_capacity: 0/5
pool_id: 22, unc_max: 3.5879, pool ACC: 0/3, pool_capacity: 3/5
pool_id: 23, unc_max: 2.1465, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 24, unc_max: 3.4258, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 25, unc_max: 3.5254, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 26, unc_max: 1.9414, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 27, unc_max: 3.3262, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 28, unc_max: 4.1914, pool ACC: 1/4, pool_capacity: 4/5
pool_id: 29, unc_max: 2.0898, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 30, unc_max: 1.0674, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 31, unc_max: 3.6523, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 32, unc_max: 3.3008, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 33, unc_max: 2.8945, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 34, unc_max: 3.4453, pool ACC: 1/1, pool_capacity: 1/5
pool_id: 35, unc_max: 3.4336, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 36, unc_max: 3.0117, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 37, unc_max: 3.5176, pool ACC: 2/3, pool_capacity: 3/5
pool_id: 38, unc_max: 3.7109, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 39, unc_max: 2.1582, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 40, unc_max: 2.7441, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 41, unc_max: 2.3828, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 42, unc_max: 0.0000, pool ACC: 0/0, pool_capacity: 0/5
pool_id: 43, unc_max: 2.9180, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 44, unc_max: 3.1680, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 45, unc_max: 2.9004, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 46, unc_max: 3.7988, pool ACC: 0/3, pool_capacity: 3/5
====> overall pools ACC: 144/196 = 0.7346938848495483"""

text_after_iter1="""
pool_id: 0, unc_max: 3.6621, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 1, unc_max: 3.2949, pool ACC: 1/5, pool_capacity: 5/5
pool_id: 2, unc_max: 2.5059, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 3, unc_max: 1.9092, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 4, unc_max: 0.0000, pool ACC: 0/0, pool_capacity: 0/5
pool_id: 5, unc_max: 1.7373, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 6, unc_max: 3.7559, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 7, unc_max: 3.3262, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 8, unc_max: 3.1230, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 9, unc_max: 2.3516, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 10, unc_max: 3.5391, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 11, unc_max: 3.7441, pool ACC: 0/1, pool_capacity: 1/5
pool_id: 12, unc_max: 3.3770, pool ACC: 1/5, pool_capacity: 5/5
pool_id: 13, unc_max: 2.1523, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 14, unc_max: 2.8086, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 15, unc_max: 3.8340, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 16, unc_max: 2.4121, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 17, unc_max: 3.6680, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 18, unc_max: 2.6641, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 19, unc_max: 3.3848, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 20, unc_max: 0.6353, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 21, unc_max: 0.0000, pool ACC: 0/0, pool_capacity: 0/5
pool_id: 22, unc_max: 3.5391, pool ACC: 0/5, pool_capacity: 5/5
pool_id: 23, unc_max: 2.1465, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 24, unc_max: 3.3633, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 25, unc_max: 3.4492, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 26, unc_max: 1.9414, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 27, unc_max: 3.2266, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 28, unc_max: 3.8496, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 29, unc_max: 2.0898, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 30, unc_max: 1.0674, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 31, unc_max: 3.5938, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 32, unc_max: 3.3008, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 33, unc_max: 2.8945, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 34, unc_max: 3.8184, pool ACC: 4/4, pool_capacity: 4/5
pool_id: 35, unc_max: 3.3867, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 36, unc_max: 3.0117, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 37, unc_max: 3.3965, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 38, unc_max: 3.6836, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 39, unc_max: 2.1582, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 40, unc_max: 2.7441, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 41, unc_max: 2.3809, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 42, unc_max: 3.7617, pool ACC: 0/1, pool_capacity: 1/5
pool_id: 43, unc_max: 2.9180, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 44, unc_max: 3.0664, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 45, unc_max: 2.8887, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 46, unc_max: 3.7852, pool ACC: 0/5, pool_capacity: 5/5
====> overall pools ACC: 150/216 = 0.6944444179534912"""

text_after_iter2 = """
pool_id: 0, unc_max: 3.6074, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 1, unc_max: 3.2949, pool ACC: 1/5, pool_capacity: 5/5
pool_id: 2, unc_max: 2.5059, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 3, unc_max: 1.9092, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 4, unc_max: 0.0000, pool ACC: 0/0, pool_capacity: 0/5
pool_id: 5, unc_max: 1.7373, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 6, unc_max: 3.7559, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 7, unc_max: 3.3262, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 8, unc_max: 3.1230, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 9, unc_max: 2.3516, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 10, unc_max: 3.5117, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 11, unc_max: 3.8438, pool ACC: 0/4, pool_capacity: 4/5
pool_id: 12, unc_max: 3.3770, pool ACC: 1/5, pool_capacity: 5/5
pool_id: 13, unc_max: 2.1523, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 14, unc_max: 2.8086, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 15, unc_max: 3.6953, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 16, unc_max: 2.4121, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 17, unc_max: 3.6680, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 18, unc_max: 2.6641, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 19, unc_max: 3.3418, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 20, unc_max: 0.6353, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 21, unc_max: 3.8086, pool ACC: 0/5, pool_capacity: 5/5
pool_id: 22, unc_max: 3.5391, pool ACC: 0/5, pool_capacity: 5/5
pool_id: 23, unc_max: 2.1465, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 24, unc_max: 3.3633, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 25, unc_max: 3.4492, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 26, unc_max: 1.9414, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 27, unc_max: 3.2266, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 28, unc_max: 3.6445, pool ACC: 1/5, pool_capacity: 5/5
pool_id: 29, unc_max: 2.0898, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 30, unc_max: 1.0674, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 31, unc_max: 3.5508, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 32, unc_max: 3.3008, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 33, unc_max: 2.8945, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 34, unc_max: 3.8184, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 35, unc_max: 3.3867, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 36, unc_max: 3.0117, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 37, unc_max: 3.3184, pool ACC: 1/5, pool_capacity: 5/5
pool_id: 38, unc_max: 3.6777, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 39, unc_max: 2.1582, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 40, unc_max: 2.7441, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 41, unc_max: 2.3809, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 42, unc_max: 4.0938, pool ACC: 1/2, pool_capacity: 2/5
pool_id: 43, unc_max: 2.9180, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 44, unc_max: 3.0527, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 45, unc_max: 2.8887, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 46, unc_max: 3.7070, pool ACC: 0/5, pool_capacity: 5/5
====> overall pools ACC: 148/226 = 0.6548672318458557
"""

# text_after_iter3 = ...



import re

def parse_pool_info(text):
    pool_info = {}
    lines = text.split('\n')
    for line in lines:
        if 'pool_id' in line:
            pool_id = int(re.search(r'pool_id: (\d+)', line).group(1))
            pool_acc = tuple(map(int, re.search(r'pool ACC: (\d+)/(\d+)', line).groups()))
            pool_info[pool_id] = pool_acc
    return pool_info

def compare_pool_info(info_before, info_after):
    changes = {}
    for pool_id, acc_before in info_before.items():
        acc_after = info_after.get(pool_id)
        if acc_after is not None and acc_before != acc_after:
            changes[pool_id] = (acc_before, acc_after)
    return changes

def total_acc(info):
    total_numerator = sum(acc[0] for acc in info.values())
    total_denominator = sum(acc[1] for acc in info.values())
    return total_numerator, total_denominator

def track_pool_info(text_info):
    pool_info = {}
    for text in text_info:
        info = parse_pool_info(text)
        for pool_id, acc in info.items():
            if pool_id not in pool_info:
                pool_info[pool_id] = []
            pool_info[pool_id].append(acc)
    return pool_info

def print_pool_info(pool_info):
    for pool_id, accs in pool_info.items():
        # Check if all ACC values are the same
        if all(acc == accs[0] for acc in accs):
            continue  # Skip this pool if there are no changes
        accs_str = ' --> '.join([f'{acc[0]}/{acc[1]}' for acc in accs])
        print(f'Pool {pool_id} ACC changed: {accs_str}')

def print_total_acc(text_info):
    total_accs = [total_acc(parse_pool_info(text)) for text in text_info]
    total_accs_str = ' --> '.join([f'{acc[0]}/{acc[1]}' for acc in total_accs])
    print(f'Total ACC changed: {total_accs_str}')

text_info = [text_before, text_after_iter1, text_after_iter2]

# Track and print the pool information
pool_info = track_pool_info(text_info)
print_pool_info(pool_info)
# Print the total ACC changes
print_total_acc(text_info)
```

```python
text_before="""
pool_id: 0, unc_max: 3.8594, pool ACC: 2/2, pool_capacity: 2/5
pool_id: 1, unc_max: 3.4570, pool ACC: 1/4, pool_capacity: 4/5
pool_id: 2, unc_max: 2.5059, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 3, unc_max: 1.9092, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 4, unc_max: 0.0000, pool ACC: 0/0, pool_capacity: 0/5
pool_id: 5, unc_max: 1.7373, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 6, unc_max: 3.7695, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 7, unc_max: 3.3262, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 8, unc_max: 3.1289, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 9, unc_max: 2.3516, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 10, unc_max: 3.7070, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 11, unc_max: 0.0000, pool ACC: 0/0, pool_capacity: 0/5
pool_id: 12, unc_max: 3.6855, pool ACC: 1/5, pool_capacity: 5/5
pool_id: 13, unc_max: 2.1523, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 14, unc_max: 2.8086, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 15, unc_max: 3.6953, pool ACC: 1/1, pool_capacity: 1/5
pool_id: 16, unc_max: 2.6270, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 17, unc_max: 3.7637, pool ACC: 1/5, pool_capacity: 5/5
pool_id: 18, unc_max: 2.6641, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 19, unc_max: 3.4707, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 20, unc_max: 0.6353, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 21, unc_max: 0.0000, pool ACC: 0/0, pool_capacity: 0/5
pool_id: 22, unc_max: 3.5879, pool ACC: 0/3, pool_capacity: 3/5
pool_id: 23, unc_max: 2.1465, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 24, unc_max: 3.4258, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 25, unc_max: 3.5254, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 26, unc_max: 1.9414, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 27, unc_max: 3.3262, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 28, unc_max: 4.1914, pool ACC: 1/4, pool_capacity: 4/5
pool_id: 29, unc_max: 2.0898, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 30, unc_max: 1.0674, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 31, unc_max: 3.6523, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 32, unc_max: 3.3008, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 33, unc_max: 2.8945, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 34, unc_max: 3.4453, pool ACC: 1/1, pool_capacity: 1/5
pool_id: 35, unc_max: 3.4336, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 36, unc_max: 3.0117, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 37, unc_max: 3.5176, pool ACC: 2/3, pool_capacity: 3/5
pool_id: 38, unc_max: 3.7109, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 39, unc_max: 2.1582, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 40, unc_max: 2.7441, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 41, unc_max: 2.3828, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 42, unc_max: 0.0000, pool ACC: 0/0, pool_capacity: 0/5
pool_id: 43, unc_max: 2.9180, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 44, unc_max: 3.1680, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 45, unc_max: 2.9004, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 46, unc_max: 3.7988, pool ACC: 0/3, pool_capacity: 3/5
====> overall pools ACC: 144/196 = 0.7346938848495483"""

text_after_iter1="""
pool_id: 0, unc_max: 3.6621, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 1, unc_max: 3.2949, pool ACC: 1/5, pool_capacity: 5/5
pool_id: 2, unc_max: 2.5059, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 3, unc_max: 1.9092, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 4, unc_max: 0.0000, pool ACC: 0/0, pool_capacity: 0/5
pool_id: 5, unc_max: 1.7373, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 6, unc_max: 3.7559, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 7, unc_max: 3.3262, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 8, unc_max: 3.1230, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 9, unc_max: 2.3516, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 10, unc_max: 3.5391, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 11, unc_max: 3.7441, pool ACC: 0/1, pool_capacity: 1/5
pool_id: 12, unc_max: 3.3770, pool ACC: 1/5, pool_capacity: 5/5
pool_id: 13, unc_max: 2.1523, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 14, unc_max: 2.8086, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 15, unc_max: 3.8340, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 16, unc_max: 2.4121, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 17, unc_max: 3.6680, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 18, unc_max: 2.6641, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 19, unc_max: 3.3848, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 20, unc_max: 0.6353, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 21, unc_max: 0.0000, pool ACC: 0/0, pool_capacity: 0/5
pool_id: 22, unc_max: 3.5391, pool ACC: 0/5, pool_capacity: 5/5
pool_id: 23, unc_max: 2.1465, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 24, unc_max: 3.3633, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 25, unc_max: 3.4492, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 26, unc_max: 1.9414, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 27, unc_max: 3.2266, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 28, unc_max: 3.8496, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 29, unc_max: 2.0898, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 30, unc_max: 1.0674, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 31, unc_max: 3.5938, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 32, unc_max: 3.3008, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 33, unc_max: 2.8945, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 34, unc_max: 3.8184, pool ACC: 4/4, pool_capacity: 4/5
pool_id: 35, unc_max: 3.3867, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 36, unc_max: 3.0117, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 37, unc_max: 3.3965, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 38, unc_max: 3.6836, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 39, unc_max: 2.1582, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 40, unc_max: 2.7441, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 41, unc_max: 2.3809, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 42, unc_max: 3.7617, pool ACC: 0/1, pool_capacity: 1/5
pool_id: 43, unc_max: 2.9180, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 44, unc_max: 3.0664, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 45, unc_max: 2.8887, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 46, unc_max: 3.7852, pool ACC: 0/5, pool_capacity: 5/5
====> overall pools ACC: 150/216 = 0.6944444179534912"""

text_after_iter2 = """
pool_id: 0, unc_max: 3.6074, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 1, unc_max: 3.2949, pool ACC: 1/5, pool_capacity: 5/5
pool_id: 2, unc_max: 2.5059, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 3, unc_max: 1.9092, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 4, unc_max: 0.0000, pool ACC: 0/0, pool_capacity: 0/5
pool_id: 5, unc_max: 1.7373, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 6, unc_max: 3.7559, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 7, unc_max: 3.3262, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 8, unc_max: 3.1230, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 9, unc_max: 2.3516, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 10, unc_max: 3.5117, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 11, unc_max: 3.8438, pool ACC: 0/4, pool_capacity: 4/5
pool_id: 12, unc_max: 3.3770, pool ACC: 1/5, pool_capacity: 5/5
pool_id: 13, unc_max: 2.1523, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 14, unc_max: 2.8086, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 15, unc_max: 3.6953, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 16, unc_max: 2.4121, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 17, unc_max: 3.6680, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 18, unc_max: 2.6641, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 19, unc_max: 3.3418, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 20, unc_max: 0.6353, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 21, unc_max: 3.8086, pool ACC: 0/5, pool_capacity: 5/5
pool_id: 22, unc_max: 3.5391, pool ACC: 0/5, pool_capacity: 5/5
pool_id: 23, unc_max: 2.1465, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 24, unc_max: 3.3633, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 25, unc_max: 3.4492, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 26, unc_max: 1.9414, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 27, unc_max: 3.2266, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 28, unc_max: 3.6445, pool ACC: 1/5, pool_capacity: 5/5
pool_id: 29, unc_max: 2.0898, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 30, unc_max: 1.0674, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 31, unc_max: 3.5508, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 32, unc_max: 3.3008, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 33, unc_max: 2.8945, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 34, unc_max: 3.8184, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 35, unc_max: 3.3867, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 36, unc_max: 3.0117, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 37, unc_max: 3.3184, pool ACC: 1/5, pool_capacity: 5/5
pool_id: 38, unc_max: 3.6777, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 39, unc_max: 2.1582, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 40, unc_max: 2.7441, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 41, unc_max: 2.3809, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 42, unc_max: 4.0938, pool ACC: 1/2, pool_capacity: 2/5
pool_id: 43, unc_max: 2.9180, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 44, unc_max: 3.0527, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 45, unc_max: 2.8887, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 46, unc_max: 3.7070, pool ACC: 0/5, pool_capacity: 5/5
====> overall pools ACC: 148/226 = 0.6548672318458557
"""

text_after_iter3 = """
pool_id: 0, unc_max: 3.6074, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 1, unc_max: 3.2949, pool ACC: 1/5, pool_capacity: 5/5
pool_id: 2, unc_max: 2.5059, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 3, unc_max: 1.9092, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 4, unc_max: 0.0000, pool ACC: 0/0, pool_capacity: 0/5
pool_id: 5, unc_max: 1.7373, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 6, unc_max: 3.7402, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 7, unc_max: 3.3262, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 8, unc_max: 3.1230, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 9, unc_max: 2.3516, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 10, unc_max: 3.4355, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 11, unc_max: 3.9004, pool ACC: 0/5, pool_capacity: 5/5
pool_id: 12, unc_max: 3.3770, pool ACC: 1/5, pool_capacity: 5/5
pool_id: 13, unc_max: 2.1523, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 14, unc_max: 2.8086, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 15, unc_max: 3.6953, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 16, unc_max: 2.4121, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 17, unc_max: 3.6309, pool ACC: 1/5, pool_capacity: 5/5
pool_id: 18, unc_max: 2.6641, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 19, unc_max: 3.3418, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 20, unc_max: 0.6353, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 21, unc_max: 3.7734, pool ACC: 0/5, pool_capacity: 5/5
pool_id: 22, unc_max: 3.5312, pool ACC: 0/5, pool_capacity: 5/5
pool_id: 23, unc_max: 2.1465, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 24, unc_max: 3.3633, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 25, unc_max: 3.4492, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 26, unc_max: 1.9414, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 27, unc_max: 3.2266, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 28, unc_max: 3.6074, pool ACC: 1/5, pool_capacity: 5/5
pool_id: 29, unc_max: 2.0898, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 30, unc_max: 1.0674, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 31, unc_max: 3.5508, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 32, unc_max: 3.3008, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 33, unc_max: 2.8945, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 34, unc_max: 3.6973, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 35, unc_max: 3.3867, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 36, unc_max: 3.0117, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 37, unc_max: 3.3184, pool ACC: 1/5, pool_capacity: 5/5
pool_id: 38, unc_max: 3.6543, pool ACC: 2/5, pool_capacity: 5/5
pool_id: 39, unc_max: 2.1582, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 40, unc_max: 2.7441, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 41, unc_max: 2.3809, pool ACC: 4/5, pool_capacity: 5/5
pool_id: 42, unc_max: 3.8379, pool ACC: 0/5, pool_capacity: 5/5
pool_id: 43, unc_max: 2.9180, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 44, unc_max: 3.0527, pool ACC: 3/5, pool_capacity: 5/5
pool_id: 45, unc_max: 2.8887, pool ACC: 5/5, pool_capacity: 5/5
pool_id: 46, unc_max: 3.7070, pool ACC: 0/5, pool_capacity: 5/5
====> overall pools ACC: 141/230 = 0.613043487071991"""



import re

def parse_pool_info(text):
    pool_info = {}
    lines = text.split('\n')
    for line in lines:
        if 'pool_id' in line:
            pool_id = int(re.search(r'pool_id: (\d+)', line).group(1))
            pool_acc = tuple(map(int, re.search(r'pool ACC: (\d+)/(\d+)', line).groups()))
            pool_info[pool_id] = pool_acc
    return pool_info

def compare_pool_info(info_before, info_after):
    changes = {}
    for pool_id, acc_before in info_before.items():
        acc_after = info_after.get(pool_id)
        if acc_after is not None and acc_before != acc_after:
            changes[pool_id] = (acc_before, acc_after)
    return changes

def total_acc(info):
    total_numerator = sum(acc[0] for acc in info.values())
    total_denominator = sum(acc[1] for acc in info.values())
    return total_numerator, total_denominator

def track_pool_info(text_info):
    pool_info = {}
    for text in text_info:
        info = parse_pool_info(text)
        for pool_id, acc in info.items():
            if pool_id not in pool_info:
                pool_info[pool_id] = []
            pool_info[pool_id].append(acc)
    return pool_info

def print_pool_info(pool_info):
    for pool_id, accs in pool_info.items():
        # Check if all ACC values are the same
        if all(acc == accs[0] for acc in accs):
            continue  # Skip this pool if there are no changes
        accs_str = ' --> '.join([f'{acc[0]}/{acc[1]}' for acc in accs])
        print(f'Pool {pool_id} ACC changed: {accs_str}')

def print_total_acc(text_info):
    total_accs = [total_acc(parse_pool_info(text)) for text in text_info]
    total_accs_str = ' --> '.join([f'{acc[0]}/{acc[1]}' for acc in total_accs])
    print(f'Total ACC changed: {total_accs_str}')

text_info = [text_before, text_after_iter1, text_after_iter2]

# Track and print the pool information
pool_info = track_pool_info(text_info)
print_pool_info(pool_info)
# Print the total ACC changes
print_total_acc(text_info)
```
